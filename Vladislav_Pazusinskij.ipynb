                                            Tabako gaiminių rūkančių Lietuvos gyventojų statistika 2014, 2019 m.

```python
smoke = pd.read_csv("RukymasLT.csv")
smoke = smoke.drop(columns ="Patikslinimas")
smoke
```

```python
smoke1 = pd.read_csv("rukymo_periodas.csv")
smoke1 = smoke1.drop(columns ="Patikslinimas")
smoke1
```

```python
period = smoke1.groupby(["Lytis","Laikotarpis","Periodas"])["Reikšmė"].mean().unstack()
```

```python
ax = period.plot(kind = "bar")

ax.set_xlabel('X ašis', fontsize=12)
ax.set_ylabel('Proc. Reikšmė', fontsize=12)
ax.set_title('Gyventojų tabako gaminių rūkymo įpročiai pagal lytį 2014 ir 2019 m. ', fontsize=12)
ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)
ax.legend(loc='best')
plt.show()
```

Beveik kas ketvirtas (24 proc.) gyventojas rūkė. Rūkymas labiau paplitęs tarp vyrų nei tarp moterų. 

```python
smoke_age=smoke.query("Periodas=='kasdien'").groupby(["Amžiaus grupės","Laikotarpis"])["Reikšmė"].mean().unstack()
```

```python
smoke_male=smoke.query("Periodas=='kasdien' and Lytis =='Vyrai'").groupby(["Amžiaus grupės","Laikotarpis"])["Reikšmė"].mean().unstack()
```

```python
smoke_female=smoke.query("Periodas=='kasdien' and Lytis =='Moterys'").groupby(["Amžiaus grupės","Laikotarpis"])["Reikšmė"].mean().unstack()
```

```python
ax = smoke_age.plot(kind = "bar")

ax.set_xlabel('Amžiaus grupės', fontsize=12)
ax.set_ylabel('Proc.reikšmė', fontsize=12)
ax.set_title('Kasdien tabako gaminius rūkančių gyventojų dalis pagal amžiaus grupes 2014, 2019 m.', fontsize=12)
ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)
ax.legend(loc='best')
plt.show()
```

```python
ax = smoke_male.plot(kind = "bar")
ax1 = smoke_female.plot(kind= "bar")

ax.set_xlabel('Vyrų amžiaus grupės', fontsize=12)
ax1.set_xlabel('Moters amžiaus grupės', fontsize=12)
ax.set_ylabel('Proc.reikšmė', fontsize=12)
ax1.set_ylabel('Proc.reikšmė', fontsize=12)
ax.set_title('Kasdien tabako gaminius rūkančių gyventojų dalis pagal amžiaus grupes vyrai 2014, 2019 m ', fontsize=12)
ax1.set_title('Kasdien tabako gaminius rūkančių gyventojų dalis pagal amžiaus grupes moterys 2014, 2019 m ', fontsize=12)
ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)
ax1.tick_params(axis='x', labelsize=12)
ax1.tick_params(axis='y', labelsize=12)
ax.legend(loc='best')
ax1.legend(loc='best')
plt.show()
```

25–34 metų amžiaus grupės kasdien rūkančių vyrų dalis, palyginti su 2014 m., sumažėjo nuo 42 proc. iki 29 proc., 25-34 moterų – atitinkamai nuo 16 proc. iki 11 proc. Tačiau rūkančių moteru dalis 
amžiaus nuo 35 iki 74 truputį padidėjo.

 

                            Elektroninių cigarečių arba panašių elektroninių prietaisų rūkymo statisktika 2019 m.

```python
el_smoke = pd.read_csv("el.cigaretes.csv")
el_smoke = el_smoke.drop(columns ="Patikslinimas")
el_smoke
```

```python
el_cig= el_smoke.groupby(["Lytis","Rūkymo dažnumas"])["Reikšmė"].mean().unstack()
```

```python
ax = el_cig.plot(kind="bar")

ax.set_xlabel('Lytis', fontsize=12)
ax.set_ylabel('Proc.reikšmė', fontsize=12)
ax.set_title('Elektroninių cigarečių arba panašių elektroninių prietaisų rūkymo įpročiai pagal lytį 2019 m. ', fontsize=12)
ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)
ax.legend(loc=(1.05,0.75))
plt.show()
```

Pagal statistika galima teigti, kad dauguma apklaustų žmonių niekada nerūkė elektroninių cigarečių. Kasdien  rūkė 3,3 proc gyventoju (2,5 proc. vyrai, 0,8 proc. moterys ), kartais rūkė 3,7 proc. (2,5 proc. vyrai 1,2 proc. moterys) , rūkančių praeityje, bet jau neberūke 6,0 proc (4,8 proc. vyrai, 1,2 proc. moterys)

                                                                               Išvada

Pagal OSP statistiką galima teigti, kad:
- Duomenys rodo, kad 2014–2019 m. sumažėjo tiek vyrų, tiek moterų rūkymas. Tačiau vyrų rūkymas vis dar yra gana aukštas, palyginti su moterimis.

- Statistikoje taip pat pateikiama informacija apie amžiaus grupes, kurios labiausiai kenčia nuo rūkymo. Didžiausias rūkymo rodiklis buvo tarp 25-34 metų amžiaus vyrų ir 35-44 metų moterų.
